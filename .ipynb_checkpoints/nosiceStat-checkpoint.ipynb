# Statistická práce o hudebních nosičích v USA 1973-2019
Hudební nosiče historicky prošly velkou obměnou. Od prvních záznamů na speciálních válečcích a prvních šelakových deskách až po digitální streamování hudby.

Dovolil bych si začít vykreslením dat prodejů hudebních nosičů ve Spojených státech amerických od roku 1973 do roku 2019.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'C:/Users/Pc.DESKTOP-U953URB/Desktop/stat/onlyUnitsUSA.csv'  # Update to the correct path
data = pd.read_csv(file_path)

clean_data = data.dropna(subset=['value_actual']).copy()
clean_data.loc[:, 'value_actual'] = pd.to_numeric(clean_data['value_actual'], errors='coerce')

pivot_data = clean_data.pivot(index='year', columns='format', values='value_actual')

# GRAF 1, plot prodejů 1973-2019 
plt.figure(figsize=(14, 8))

# Jaké formáty?
formats_to_plot = ['LP/EP', 'Vinyl Single', '8 - Track', 'Cassette', 'CD', 'Download Single']
for fmt in formats_to_plot:
    if fmt in pivot_data.columns:
        plt.plot(pivot_data.index, pivot_data[fmt], label=fmt)

plt.title('GRAF 1: Prodeje všech hudebních nosičů v USA v letech 1973-2019 (v milionech)')
plt.xlabel('Rok')
plt.ylabel('Prodeje (mil)')
plt.legend()
plt.grid(True)
plt.show()

# GRAF 2, plot prodejů 1973-1990
plt.figure(figsize=(14, 8))

lp_ep_1973_1990 = pivot_data.loc[1973:1990, 'LP/EP']

plt.plot(lp_ep_1973_1990.index, lp_ep_1973_1990, label='LP/EP')

for fmt in formats_to_plot:
    if fmt in pivot_data.columns:
        plt.plot(pivot_data.loc[1973:1990].index, pivot_data.loc[1973:1990, fmt], label=fmt)

plt.title('GRAF 2: Prodeje hudebních nosičů v detailu (1973 - 1990)')
plt.xlabel('Rok')
plt.ylabel('Prodeje (mil)')
plt.yticks(range(0, 501, 50))  # Nastaveno na 50 pro přehlednost
plt.legend()
plt.grid(True)
plt.show()

# GRAF 3, plot prodejů 1990-2019
plt.figure(figsize=(14, 8))

lp_ep_1990_2019 = pivot_data.loc[1990:2019, 'LP/EP']

plt.plot(lp_ep_1990_2019.index, lp_ep_1990_2019, label='LP/EP')

formats_to_plot = ['Vinyl Single', '8 - Track', 'Cassette', 'CD']
for fmt in formats_to_plot:
    if fmt in pivot_data.columns:
        plt.plot(pivot_data.loc[1990:2019].index, pivot_data.loc[1990:2019, fmt], label=fmt)

plt.title('GRAF 3: Prodej hudebních nosičů v USA v letech 1990-2019 se zaměřením na LP/kazety/CD')
plt.xlabel('Rok')
plt.ylabel('Prodeje (mil)')
plt.ylim(0, 50)
plt.yticks(range(0, 51, 5))
plt.legend()
plt.grid(True)
plt.show()
```

## Poznámky ke grafu

Dalo by se říct, že vinylové desky byly dle grafu jako formát"zabity" kazetami a CD (na variantu CD vs. LP se hodlám podívat viz níže). Kazety byly levnější a šly jednoduše přenahrát. Oproti tomu druhý zmíněný formát byl dražší a měl nepřepisovatelný obsah. 90. léta však, i jak graf vypovídá znamenala výhru CD nad kazetami.

Můžeme avšak pozorovat, že při příchodu CD byla křivka nárustu prodejů o dost strmější než tomu bylo právě u zmíněných kazet.

### GRAF 3
Grafem 3 jsem chtěl ukazát nárůst prodeje vinylů od roku 2005, který se začíná pomalu přibližovat k prodejům CD vyznačeným fialovou barvou v roce 2019. Rád bych zde doplnil i aktuální informaci, že v roce 2023 byly po mnoha desitiletích prodejů vinylových desek vyšší než prodeje CD ([zdroj](https://www.popsci.com/technology/why-vinyl-records-are-better-than-cds/)). 


## Hypotéza:
Nulová hypotéza (H₀): __Prodeje CD nemají významný negativní vliv na prodej LP/EP__.

Alternativní hypotéza (H₁): __"CDčka zabila Vinyly" (Prodeje CD mají významný negativní vliv na prodej LP/EP)__.

Ověření hodlám provést přes lineární regresi:

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import statsmodels.api as sm

file_path = 'C:/Users/Pc.DESKTOP-U953URB/Desktop/stat/onlyUnitsUSA.csv'
data = pd.read_csv(file_path)

#reshape dat
pivot_data = data.pivot_table(index='year', columns='format', values='value_actual', aggfunc='sum')

#filter, kde záznamy CD nejsou
clean_data = pivot_data.dropna(subset=['CD', 'LP/EP'])

X = clean_data['CD'].values.reshape(-1, 1)  # CD sales (nezávislá proměnná)
y = clean_data['LP/EP'].values  # LP/EP sales (závislá proměnná)

# Vytvoření a trénování lineárního regresního modelu
model = LinearRegression()
model.fit(X, y)

# Získání sklonu (beta_1) a interceptu (beta_0)
slope = model.coef_[0]
intercept = model.intercept_

# Predikce prodejů LP/EP pomocí modelu
y_pred = model.predict(X)

# Vykreslení dat a regresní přímky
plt.scatter(X, y, color='blue', label='Skutečné prodeje LP/EP')
plt.plot(X, y_pred, color='red', label='Regresní přímka (prodeje CD vs LP/EP)')
plt.xlabel('Prodeje CD (v milionech jednotek)')
plt.ylabel('Prodeje LP/EP (v milionech jednotek)')
plt.title('Lineární regrese: Prodeje CD vs Prodeje LP/EP')
plt.legend()
plt.grid(True)
plt.show()

#odkomentovat pro vyzkoušení běhu
#print(f"Slope (Beta_1): {slope}")
#print(f"Intercept (Beta_0): {intercept}")

r_squared = model.score(X, y)
#print(f"R-squared: {r_squared}")

# if slope < 0:
#     print("Negative relationship")
# else:
#     print("Positive relationship")

#OLS regresní analýza
# Přidáme konstantu pro intercept
#X_with_intercept = sm.add_constant(X)

# Vytvoříme model
#model = sm.OLS(y, X_with_intercept)
#results = model.fit()

# Výstup výsledků
#print(results.summary())
```


Na základě výpočtu lineární regrese jsme získali následující výsledky:

__Koeficient sklonu (Beta₁ = -0.244)__:
        Tento koeficient nám říká, že s __každým nárůstem prodeje CD o milion jednotek se prodej LP/EP sníží o 0.244 milionu jednotek__.
        Tento negativní koeficient potvrzuje, že __existuje negativní vztah mezi prodeji CD a LP/EP__, což znamená, že jak rostly prodeje CD, prodeje LP/EP klesaly. Tento __výsledek podporuje hypotézu, že nárůst popularity CD formátu měl negativní dopad na LP/EP__.

__Intercept (Beta₀ = 161.99)__:
        Intercept (průsečík) vyjadřuje, že pokud by prodeje CD byly nulové, očekávali bychom prodeje LP/EP kolem 161.99 milionu jednotek. I když to může být čistě teoretický odhad, poskytuje to základní představu o tom, jaký by byl prodej LP/EP, pokud by CD neexistovaly.

__R-squared (R² = 0.432)__:
        Tento koeficient nám říká, že __43.2 % variability v prodejích LP/EP může být vysvětleno prodeji CD__. Jinými slovy, přibližně 43 % změn v prodejích LP/EP lze přičíst změnám v prodeji CD.
        Ačkoli tento koeficient naznačuje středně silnou korelaci, není dostatečně vysoký na to, aby vysvětlil veškeré změny v prodejích LP/EP. To znamená, že kromě CD mohly k poklesu prodeje LP/EP přispět i další faktory (například kazety, digitalizace, změny ve spotřebitelských preferencích nebo příchod (jiných) nových hudebních formátů).

## Závěr:

Na základě těchto výsledků (p-hodnota dle OLS regresivní analýzy je menší než hladiny významnosti 0.05) můžeme __zamítnout nulovou hypotézu (H₀)__ a __přijmout alternativní hypotézu (H₁)__, která tvrdí, že prodeje CD měly významný negativní vliv na prodej LP/EP. __Záporný sklon lineární regrese (-0.244) indikuje, že nárůst prodeje CD vedl k poklesu prodeje LP/EP__.

Nicméně, vzhledem k tomu, že R-squared je relativně středně vysoký (0.432), lze říci, že prodeje CD nevysvětlují veškerý pokles LP/EP prodejů. Je pravděpodobné, že do poklesu prodeje LP/EP zasáhly i další faktory, jako byly kazety, nárůst digitálních formátů a měnící se hudební trh. To můžeme vidět i na grafu 1. 

Tuto hypotézu jsem však zformuloval hlavně proto, že je to jedno z největších témat sběratelů a "historiků" zabývající se vývojem hudebních nosičů. V potaz můžeme vzít i přenosnost CD proti LP. Také začaly být do aut implementovány přehrávače CD (předtím i kazet), což u vinylových desek z fyzikálních důvodu zkrátka nešlo a nejde doteď. Vinylové desky (viz Graf 3 a dodatečné informace k němu) se však vracejí a stávají se poměrně velkým trendem.
